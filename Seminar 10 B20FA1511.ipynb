```python
#data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,10)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = plt.axes
```

```python
#Даалгавар 1
plt.plot(x,y)
plt.title('Graph 1')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Даалгавар 2
figure = plt.figure()

axes = figure.add_axes([0.0, 0.0, 0.9, 0.9])
axes2 = figure.add_axes([0.07, 0.55, 0.35, 0.3])

axes.plot(x,y,color='orange')
axes2.plot(x,y)
```

```python
#Даалгавар 3
plt.subplot(2,2,1)
plt.plot(x,y, 'b--')

plt.subplot(2,2,2)
plt.plot(x,z, 'r')
```

```python
#Даалгавар 4
plt.subplot(2,2,1)
plt.plot(x,y, 'b--')

plt.subplot(2,2,2)
plt.plot(x,z, 'r')

plt.subplot(2,2,3)
plt.plot(x,y, 'b--')

plt.subplot(2,2,4)
plt.plot(x,z, 'r')
```
